```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x=[1,2,3,4,5]
y=[10,20,30,40,50]
plt.plot(x,y)
plt.title("Line Plot")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.show()
```

```python
subjects=['Math','Science','English']
marks=[80, 90, 75]
plt.bar(subjects,marks)
plt.title("Bar Chart")
plt.xlabel("Subjects")
plt.ylabel("Marks")
plt.show()
```

```python
data=[1,2,2,3,3,3,4,4,5]
plt.hist(data)
plt.title("Histogram")
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.show()
```

```python
x=[1,2,3,4,5]
y=[10,15,20,25,30]
plt.scatter(x,y)
plt.title("Scatter Plot")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.show()
```

```python
sizes=[40,30,20,10]
labels=['A','B','C','D']
plt.pie(sizes, labels=labels)
plt.title("Pie Chart")
plt.show()
```

```python
tips=sns.load_dataset("tips")
tips.head()
```

```python
sns.scatterplot(x="total_bill", y="tip", data=tips)
plt.title("Seaborn Scatter Plot")
plt.show()
```

```python
sns.histplot(tips["total_bill"])
plt.title("Seaborn Histogram")
plt.show()
```

```python
sns.boxplot(x="day",y="total_bill",data=tips)
plt.title("Seaborn Box Plot")
plt.show()
```
