```python
import pandas as pd
df = pd.read_csv("X data.csv")
df.head()
```

```python
df.info()
```

```python
df['category'].value_counts()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.countplot(x='category', data=df)
plt.title("Sentiment Distribution")
plt.xlabel("Sentiment")
plt.ylabel("Count")
plt.show()
```

```python
df['Sentiment'] = df['category'].replace({
    -1.0: 'Negative',
     0.0: 'Neutral',
     1.0: 'Positive'
})
df.head()
```

```python
sns.countplot(x='Sentiment',data=df)
plt.title("Positive, Neutral and Negative Tweets")
plt.xlabel("Sentiment")
plt.ylabel("Count")
plt.show()
```
